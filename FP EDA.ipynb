```python
import pandas as pd

# Load the dataset
df = pd.read_csv(r"C:\Users\neeti\Documents\ISB_Class of Summer_2025\04 Term 4\Foundation\wfp_food_prices_ind.csv")

# Display basic info
print(df.info())

# Preview the first few rows
print(df.head())
```

```python
# Check for missing values
missing_values = df.isnull().sum()
print(missing_values[missing_values > 0])
```

```python
# Ensure date column is in datetime format
df['date'] = pd.to_datetime(df['date'])

# Convert price to numeric (if not already)
df['price'] = pd.to_numeric(df['price'], errors='coerce')
```

```python
print("Unique Commodities:", df['commodity'].unique())
print("Unique Markets:", df['market'].unique())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Filter for only 'Retail' price type
df_retail = df[df['pricetype'] == 'Retail']

# Histogram of prices
plt.figure(figsize=(8,5))
sns.histplot(df_retail['price'], bins=50, kde=True)
plt.title("Distribution of Retail Prices")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.show()

```

```python
import matplotlib.pyplot as plt

# Filter data for only "Retail" price type
df_retail = df[df['pricetype'] == 'Retail']

# Aggregate price by month
df_monthly_retail = df_retail.groupby(df_retail['date'].dt.to_period("M"))['price'].mean()

# Plot time series
plt.figure(figsize=(12,6))
df_monthly_retail.plot()
plt.title("Average Monthly Retail Price Over Time")
plt.xlabel("Date")
plt.ylabel("Price")
plt.grid(True)
plt.show()

```
